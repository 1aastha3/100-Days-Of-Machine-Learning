# What is Machine Learning

Machine Learning is a statistical approach to make predictions and classifications. 

### There are two types of datasets you want to work with. 


#### 1- Training Data


This corresponds to that amount of your dataset which goes into training your machine to your model. 

#### 2- Testing Data

This corresponds to the remaining untrained data which is put into testing the training and your model

#### Remember: If your model fits well to the training data, doesnt meant that it fits well to your testing data. Your aim is to find a model that fits best to your testing dada.

## How to decide which set to put as training or testing dataset?

### Cross validation

1- Estimating the parameters is called training the algorithm (usually 75%) 
2- Evaluating the method is called testing the algorithm. (usually 25%)

Cross validation compares various machine learning models to help understand which model fits the best to our testing set. Cross validation also helps to include all 25% batches of the whole dataset to be counted as a testing dataset. It will find out which 25% part of the dataset is apt to be used as a testing data. 

The final outcome gives the number of times the model has gone correct or incorrect. If we apply this to all the models, then the one with the lowest incorrect is the one best for our type of data.

Types of cross validations:

1- Four fold CV = the dataset is broken into 4 parts/folds

2-  Leave one out CV = each sample is a fold

3- ten fold CV is the most commonly used.

## BIAS-VARIANCE TRADEOFF

1- Validation set approach (50-training and 50-testing) is a bad distribution of dataset because using on 50% of the data for the training can lead to missing of some interesting values and trends which will eventually fail in the testing dataset. THIS METHOD IMPLIES THAT THERE IS A HIGHER BIAS AND SMALL VARIATION. 

2- Leave one out approach (each sample is a testing dataset) is also a bad technique due to higher execution time. Our model becomes sensitive to each data point and leads to unwanted variations due to outliers. THIS METHOD HAS VERY LOW BIAS BUT HIGH VARIATION.

3- K-fold method is the best method to classify our models that suit the dataset well. Lower value of k approches to the set validation method and higher value of k brings us to the LOOCV method both of which are undesirable.

## CONFUSION MATRIX

The GREEN boxes imply the number of times that particular model has given CORRECT predictions

The RED boxes imply the ni=number of time the model has given INCORRECT predictions

HOW TO COMPARE?

1- Create a confusion matrix for all the models applied. 
2- COmpare the number of false positives and false negatives.
3- If the number is close, we use different metrics like sensitivity, specificity, ROC, AUC

There can exist nxn confusion matrix depending upon the validation parameters.

## Sensitivity and Specificity

SENSITVITY: What percentage having XYZ were correctly identified. In other words, higher the sensitivity, implies that the model is good at identifying positives.

SPECIFICITY: What percentage of not having XYZ was correctly identified. In other words, higher the specificity, implies that the model is good at identifying negatives.

## WHAT IS BIAS?

Bias is a parameter that tells us how much is our machine learning model able to depict the "true" relationship of the data. FOr example: a linear regression does not fit the data as good as the polynomial regression. This means, that linear regression has high bias and polynomial regression has lower bias

## WHAT IS VARIANCE

The difference between the fits of the model on different portion of the datasets is called variance. Linear regression doesnt fit the training set that well, but it is consistent with the testing dataset. Hence it can make good predictions because there is not much variance in the testing dataset. 


Whereas, in polynomial regression, the training data fits so well, but it become hard to estimate the fit of the model on the testing data set. It may work terrible and provide high variance. 

#### THIS IS CALLED BIAS-VARIANCE TRADEOFF
