### Data cleaning: missing handling data

```python
df['age'].fillna(value=df['age'].mean(),inplace=True)
```

```python
df['gender'].fillna(value='Bi',inplace=True)
```

```python
df['height'].fillna(value=df['height'].mean(),inplace=True)
```

```python
df['happy?'].fillna(value='N.A.', inplace=True)
```

```python
df.iloc[0:15]
```

### Data encoding

```python
df2 = pd.get_dummies(df.gender)
```

```python
df2
```

```python
df['gender'] = df['gender'].astype('category')
df['G']=df['gender'].cat.codes
df['gender']=df['G']
df.drop('G',axis=1,inplace=True)

df['height'] = df['height'].astype('category')
df['H']=df['height'].cat.codes
df['height']=df['H']
df.drop('H',axis=1,inplace=True)

df['happy?'] = df['happy?'].astype('category')
df['?']=df['happy?'].cat.codes
df['happy?']=df['?']
df.drop('?',axis=1,inplace=True)
```
